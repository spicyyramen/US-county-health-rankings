# Look at Violent Crime Rates in California

```python
# open counties location file
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)
# create animated map of violent crime rates across Year Spans
fig = px.choropleth(df2.loc[(df2['State'] == 'CA')&(df2['County']!= "California") & (df2['Measure name'] == "Violent crime rate")], 
                    geojson=counties, 
                    locations='FIPS Code', 
                    color='Raw value',
                           color_continuous_scale="turbo",
                           range_color=(0, 700),
                           scope="usa",
                            title='<b>Violent Crime Rates in California by County',
                           labels={'Raw value':'Violent Crime Rate',
                                  'fipscode':'FIPS Code'},
                            hover_name = 'County',
                    animation_frame="Year span"
                          )
fig.update_layout(margin={"r":0,"t":50,"l":0,"b":0},
                 height=600,
                 width=800)
# Adjust map geo options
fig.update_geos(showcountries=False, showcoastlines=True,
                fitbounds="locations",
                )
fig.show()
```
